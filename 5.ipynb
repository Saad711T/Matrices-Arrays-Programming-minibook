```python
import numpy as np
# التأكد من نجاح تضمين المكتبة
print("Success")
```

```python
# مصفوفة عادية
x = np.array([1, 2, 3])

# طباعة العنصر رقم 0 وهو العنصر الأول في المصفوفة
print (x[0])
```

```python
y = np.array([3, 4, 5])
# تعريف مصفوفة أخرى
```

```python
# جمع مصفوفتين

sumarrays = x+y

# الناتج مصفوفة تم جمعها بين مصفوفة اكس وواي
print(sumarrays)
```

```python
# مصفوفة ثنائية أبعاد

a = np.array([[1, 2], [3, 4]])
```

```python
print(a[0][1])
```

```python
# مصفوفة ثلاثية الأبعاد

b = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```

```python
print(b[0][1])
```

```python
# ضرب المصفوفات

# تعريف المصفوفتين
a = np.array([[1, 2], [3, 4]])
c = np.array([[5, 6], [7, 8]])
```

```python
# حفظ الضرب في متغير النتيجة وطباعته

result = np.dot(a, c)
print("Matrix multiplication result:")
print(result)
```

```python
# مصفوفة صفرية

zero = np.zeros([6])
```

```python
print(zero)
# تتكون من 6 عناصر كل عناصرها صفر
```

```python
threedzeros = np.zeros((2, 3, 4))

# 3D Array with all elements is zeros
```

```python
print(threedzeros)
```
